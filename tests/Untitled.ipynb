```python
%matplotlib widget


# Imports
from astropy.io import fits
import matplotlib.pyplot as plt
import mplcursors
import numpy as np

# Activer le mode widget de Matplotlib pour une interactivité dans Jupyter Lab

# Chemin vers votre fichier .fits
path_to_fits_file = "../data/UGC_09618_2MASS_H.fits"

# Ouvrir le fichier .fits
with fits.open(path_to_fits_file) as hdul:
    # Accéder à la première extension (HDU)
    hdu = hdul[0]
    
    # Accéder aux données de l'image
    image_data = hdu.data

    # Vérification que les données existent
    if image_data is None:
        print("Pas de données d'image dans ce fichier FITS.")
    else:
        # Créer une figure et un axe
        fig, ax = plt.subplots()

        # Afficher l'image
        img = ax.imshow(image_data, cmap='gray', origin='lower', interpolation='none')

        # Ajouter une barre de couleur
        plt.colorbar(img, ax=ax)

        # Ajouter des titres
        ax.set_title('Image from FITS file')

        # Activer la fonction de zoom
        plt.axis('on')

        # Ajouter des curseurs interactifs
        mplcursors.cursor(img, hover=True)

        # Afficher la figure
        plt.show()
```
